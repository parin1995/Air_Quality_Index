```python
import pandas as pd
import numpy as np
```

```python
dataset_name = "AQI_dataset_CO.csv"

pollutant_time_lag = 1
meteorological_time_lag = 1

pollutant_col = ['Daily Max 8-hour CO Concentration'] 

meteorological_col = ['altimeter_set_1_INHG_Median',
                      'air_temp_set_1_Fahrenheit_Median',
                      'relative_humidity_set_1_%_Median',
                      'wind_speed_set_1_Miles/hour_Median',
                      'wind_direction_set_1_Compass_Median',
                     'altimeter_set_1_INHG_High',
                      'air_temp_set_1_Fahrenheit_High',
                      'relative_humidity_set_1_%_High',
                      'wind_speed_set_1_Miles/hour_High',
                     'altimeter_set_1_INHG_Low',
                      'air_temp_set_1_Fahrenheit_Low',
                      'relative_humidity_set_1_%_Low',
                      'wind_speed_set_1_Miles/hour_Low']

'''
pollutant_col = ['Daily Max 8-hour CO Concentration',
 'Daily Max 1-hour NO2 Concentration',
 'Daily Max 8-hour Ozone Concentration', 
 'Daily Mean PM10 Concentration',
 'Daily Mean PM2.5 Concentration', 
 'Daily Max 1-hour SO2 Concentration']
'''
```

```python
dataset = pd.read_csv(dataset_name)
```

```python
dataset.shape
```

```python
dataset['date'] = pd.to_datetime(dataset['Unnamed: 0'])
dataset = dataset.drop(columns = ['Unnamed: 0'], errors = 'ignore')
```

```python
dataset
```

```python
# Monday = 0, Sunday = 6
dataset['weekend'] = dataset['date'].dt.dayofweek > 5
```

```python
type(dataset)
```

```python
import math
def degrees_to_cardinal(d):
    
    dirs = ["N", "NNE", "NE", "ENE", "E", "ESE", "SE", "SSE","S", "SSW", "SW", "WSW", "W", "WNW", "NW", "NNW"]
    ix = (d + 11.25)/22.5
    if(math.isnan(ix)):
        ix = 0
    else:
        ix = int(ix)
    return dirs[ix % 16]
```

```python
dataset['wind_direction_set_1_Compass_Median'] = dataset['wind_direction_set_1_Degrees_Median'].apply(degrees_to_cardinal)
```

```python
dataset = dataset.drop(['wind_direction_set_1_Degrees_High','wind_direction_set_1_Degrees_Median','wind_direction_set_1_Degrees_Low'], axis = 1)
```

```python
dataset.columns
```

```python
for i in range(pollutant_time_lag, 0, -1):
    print("time_lag : " + str(i))
    for j in pollutant_col:
        print("col : " + str(j))
        new_column_name = j + '_(t-%d)' % (i)
        dataset[new_column_name] = dataset[j].shift(i)
```

```python
dataset
```

```python
for i in range(meteorological_time_lag, 0, -1):
    print("time_lag : " + str(i))
    for j in meteorological_col:
        print("col : " + str(j))
        new_column_name = j + '_(t-%d)' % (i)
        dataset[new_column_name] = dataset[j].shift(i)
```

```python
n=max(pollutant_time_lag, meteorological_time_lag)
dataset = dataset[n:]
```
